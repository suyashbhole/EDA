# **concat()**

**Syntax:**

**pd.concat(objs, axis=0, join='outer', ignore_index=False)**

axis = 0  Combine vertically (row-wise) (default)

axis = 1  Combine horizontally (column-wise)

join='outer'  Keep all columns (default)

join='inner'  Keep only common columns

ignore_index=True  Reset index

ignore_index=False  Keep original index

# **concat()**

Used to append or stack DataFrames

Works along an axis (rows or columns)

Does NOT require common column names

```python
# Let us concatenate df1 and df2.
df3=pd.concat([df1,df2])# default axis=0 row-wise, default outer join
print(df3)
```

# **merge()**
merge() is used to combine DataFrames based on a common column (key)  similar to SQL JOIN.

**Syntax:**

**pd.merge(left_table, right_table, how='inner', on=None, left_on=None, right_on=None)**

on  Common column name

left_on / right_on  Different column names

```python
# merge:
df5=pd.merge(df3,df4,on="id")# default inner join
print(df5)
```

```python
df5=pd.merge(df3,df4,left_on="id",right_on="id")
print(df5)

# left_on and right_on tells to Use this column from
# the LEFT /RIGHT DataFrame for matching
# It is required because sometimes common column names
# are different in both DataFrames.
# output is same as above
```

```python
print(df1.columns)
print(df2.columns)
```

```python
# Outer Join (Union)
df4=pd.merge(df1,df2,on="id",how="outer")
print(df4)
```

```python
# Let us save above result in csv file
df4.to_csv("outer_join_df.csv")# give any name
```

```python
# Let us save the above data in an excel file
df4.to_excel("outer_join_excel_df.xlsx")
```

```python
# Inner Join (intersection)
df4=pd.merge(df1,df2,on="id",how="inner")
print(df4)
```

```python
# Right Join
df4=pd.merge(df1,df2,on="id",how="right")
print(df4)
```

```python
# Left Join
df4=pd.merge(df1,df2,on="id",how="left")
print(df4)
```

```python
df1
```

```python
# Joining Based on Index
# This merge is working like: Row 0 with Row 0, Row 1 with Row 1
# It is not matching column id in both tables

df4=pd.merge(df1,df2,left_index=True,right_index=True)
print(df4)
```

```python
# Joining Based on Index
df4=pd.merge(df1,df2,left_index=False,right_index=False)
print(df4)

# why empty dataframe?
# When you do NOT specify on=common_col_name,
# Pandas automatically tries to merge on all common column names.
# So internally it becomes: pd.merge(df1, df2, on=['id', 'val_1', 'val_2'])
# Got empty table as there are no rows where all three columns match exactly
```

```python
df4=pd.merge(df1,df2,on='id',left_index=False,right_index=False)
print(df4)
```
